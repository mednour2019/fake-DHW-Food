```python
from faker import Faker
import pandas as pd
import random

# Initialize Faker
fake = Faker()

# Number of records in each dimension
num_restaurants = 10
num_managers = 5
num_customers = 20
num_orders = 100

# Generate dim_restaurant
restaurants = []
for i in range(num_restaurants):
    restaurants.append({
        'restaurant_id': i + 1,
        'restaurant_name': fake.company(),
        'location': fake.address(),
        'cuisine_type': fake.random_element(['Italian', 'French', 'Japanese', 'Mexican', 'Indian'])
    })
dim_restaurant = pd.DataFrame(restaurants)

# Generate dim_manager
managers = []
for i in range(num_managers):
    managers.append({
        'manager_id': i + 1,
        'manager_name': fake.name(),
        'manager_email': fake.email()
    })
dim_manager = pd.DataFrame(managers)

# Generate dim_customer
customers = []
for i in range(num_customers):
    customers.append({
        'customer_id': i + 1,
        'customer_name': fake.name(),
        'customer_email': fake.email(),
        'customer_phone': fake.phone_number()
    })
dim_customer = pd.DataFrame(customers)

# Generate dim_date
dates = pd.date_range(start='2023-01-01', end='2023-12-31').to_frame(index=False, name='date')
dates['day'] = dates['date'].dt.day
dates['month'] = dates['date'].dt.month
dates['year'] = dates['date'].dt.year
dim_date = dates.reset_index(drop=True)
dim_date['date_id'] = dim_date.index + 1

# Generate fact_orders
orders = []
for i in range(num_orders):
    order_amount = round(random.uniform(10, 500), 2)
    profit = round(random.uniform(1, 100), 2)
    benefit = round(order_amount - profit, 2)
    
    orders.append({
        'order_id': i + 1,
        'restaurant_id': random.randint(1, num_restaurants),
        'manager_id': random.randint(1, num_managers),
        'customer_id': random.randint(1, num_customers),
        'date_id': random.randint(1, len(dim_date)),
        'order_amount': order_amount,
        'profit': profit,
        'benefit': benefit
    })
fact_orders = pd.DataFrame(orders)

# Save to Excel
with pd.ExcelWriter('datawarehouse_star_schema.xlsx') as writer:
    dim_restaurant.to_excel(writer, sheet_name='dim_restaurant', index=False)
    dim_manager.to_excel(writer, sheet_name='dim_manager', index=False)
    dim_customer.to_excel(writer, sheet_name='dim_customer', index=False)
    dim_date.to_excel(writer, sheet_name='dim_date', index=False)
    fact_orders.to_excel(writer, sheet_name='fact_orders', index=False)

print("Data warehouse star schema has been generated and saved to 'datawarehouse_star_schema.xlsx'")
```
